```python
df = pd.read_csv("data.csv")
# To display the top 5 rows
df.head(5)
```

```python
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
df.head(5)
```

```python
df['Transmission Type'].value_counts()
```

```python
one_hot = pd.get_dummies(df['Transmission Type']).reset_index()
```

```python
df2 = df.reset_index()
```

```python
one_hot
```

```python
df2 = pd.merge(one_hot, df2, on='index')
```

```python
df2.drop(columns='index')
```

```python
df_filter = df2[df2['UNKNOWN'] != 1]
```

```python
# look up inplace
df2 = df2.fillna(method='ffill')
```

```python
# independent variable
X = df2[{'MANUAL', 'DIRECT_DRIVE', 'AUTOMATIC', 'AUTOMATED_MANUAL', 'Engine HP'}]
# target variable
y = df2['MSRP']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, 
                                                    test_size = 0.2)
```

```python
import statsmodels.api as sm

# Adding a constant to get an intercept
X_train_sm = sm.add_constant(X_train)
```

```python
# Fitting the resgression line using 'OLS'
# ordinary lease squares
lr = sm.OLS(y_train, X_train_sm).fit()

# Printing the parameters
lr.params
```

```python
lr.summary()
```

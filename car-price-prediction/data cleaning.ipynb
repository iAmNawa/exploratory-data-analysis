```python
df['Year'].value_counts()
```

```python
# only include 2015-2017
df_2015_to_2017 = df[(df['Year']>=2015)]
```

```python
df_2015_to_2017['Year'].value_counts()
```

```python
df_2015_to_2017_engine_hp = df_2015_to_2017['HP']
df_2015_to_2017_engine_hp.hist(bins=50, grid=False)
```

```python
# remove cars below 100hp && cars above 450
df_2015_to_2017_100_to_450 = df_2015_to_2017[(df_2015_to_2017['HP'] > 100)&(df_2015_to_2017['HP'] < 450)]
```

```python
df_2015_to_2017_100_to_450['HP'].hist(bins=50, grid=False)
```

```python
df_2015_to_2017_100_to_450.min()
```

```python
# independent variable
X = df_2015_to_2017_100_to_450['HP']
# target variable
y = df_2015_to_2017_100_to_450['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, 
                                                    test_size = 0.3, random_state = 100)
```

```python
import statsmodels.api as sm

# Adding a constant to get an intercept
X_train_sm = sm.add_constant(X_train)
```

```python
# Fitting the resgression line using 'OLS'
# ordinary lease squares
lr = sm.OLS(y_train, X_train_sm).fit()

# Printing the parameters
lr.params
```

```python
lr.summary()
```

```python
# Visualizing the regression line
# high correlation makes a biased assumption
plt.scatter(X_train, y_train)
# y = c + mX
# Price = coefficient const + cooefficient HP * HP
plt.plot(X_train, -5.077e+04 + 365.1887*X_train, 'r')
plt.show()
```
